# Heat Map

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset into my program
spotify = pd.read_csv('./SpotifyWithDate.csv')

# Drop columns
spotify.drop(columns = 'track_id', inplace = True)

# Update popularity into containers
spotify["popularity"] = ((spotify["popularity"] - 1)//20) + 1

# # Draw heatmap
fig, ax = plt.subplots(figsize=(20,20))

heatmap = sns.heatmap(spotify.corr(), vmin=-1, vmax=1,annot=True,ax = ax)
```

```python
pairplot = sns.pairplot(spotify, diag_kind='kde')
```

```python
# Get heatmap and pairplot pictures
from pylab import savefig
heatmap_pic = heatmap.get_figure()    
heatmap_pic.savefig('heatmap.png', dpi=400)
pairplot.savefig('pairplot.png', dpi=400)
```
